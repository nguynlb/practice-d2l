```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt


def synthetic_data(n_samples: int, W: torch.tensor, b: int):
    X = torch.normal(0, 1, (n_samples, len(W)))
    y = torch.matmul(X, W) + b 
    y += torch.normal(0, 0.01, (n_samples,))
    return X, y.reshape((-1, 1))
```

```python
true_W = torch.tensor([3, -5]).type(torch.float)
true_b = -2
n_samples = 1000

torch.manual_seed(10)

X, y = synthetic_data(n_samples, true_W, true_b)
print(f"Shape of features: {X.shape}")
print(f"Shape of targets: {y.shape}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.scatter(X[:, 0], y)
plt.xlabel("Feature 1")
plt.ylabel("Target")

ax2.scatter(X[:, 1], y)
plt.xlabel("Feature 2")
plt.ylabel("Target")
```
